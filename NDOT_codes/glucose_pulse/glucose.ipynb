```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np



# Read CSV into panda

df = pd.read_csv("all_ion_short_ca1p32_HighKhap.csv")
# print(df1.head())



fig = plt.figure(figsize =(10.0, 7.0))

COLOR1 = '#000000'
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.size'] = 14
plt.rcParams['text.color'] = COLOR1
plt.rcParams['axes.labelcolor'] = COLOR1
plt.rcParams['xtick.color'] = COLOR1
plt.rcParams['ytick.color'] = COLOR1

# set tick width
plt.rcParams['xtick.major.size'] = 4
plt.rcParams['xtick.major.width'] = 3
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.major.size'] = 4
plt.rcParams['ytick.major.width'] = 3
plt.rcParams['ytick.direction'] = 'in'

# change all spines
for axis in ['top','bottom','left','right']:
    plt.gca().spines[axis].set_linewidth(3)
    
#plt.title("% F substitution vs PE", fontsize=36)

ions_bulk=df[df['position']==-150]
ions_film=df[df['position']==0]
ions_surface=df[df['position']==250]

plt.xlabel('t (min)', fontsize=24)

#plt.xlim(0.0, 19.6)
#plt.ylim(0.0, 300.0)
plt.ylabel(r'Conc. (mmol/L)', fontsize=24)
#plt.ylabel(r'# Particle', fontsize=24)


#plt.plot(df1.iloc[:,1], df1.iloc[:,3], color='c', linestyle='-', linewidth=2)
plt.plot(ions_bulk['Time in min'], ions_bulk['glu'], color='navy', linestyle='-', linewidth=3)
plt.plot(ions_film['Time in min'], ions_film['glu'], color='c', linestyle='-', linewidth=3)
plt.plot(ions_surface['Time in min'], ions_surface['glu'], color='orange', linestyle='-', linewidth=3)
plt.xlim(0,30)
plt.legend(['Glucose (bulk)','Glucose (Film)','Glucose (Teeth Surface)'], bbox_to_anchor=(0.95, 0.95), ncol=1)

# Show Plot
plt.tight_layout()
#plt.show()
fig.savefig("glucose-conc-time.jpg", dpi = 600)
```
